```python
#matplotlib.pyplot as plt
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])  # adds a line to the current axes
plt.title("Simple plot")
plt.show()
```

```python
# matplotlib.collections and patches examples
"""
This module defines Collection classes, which are efficient containers for drawing many similar objects at once.
Instead of drawing 100 circles individually with plt.Circle, you can use a PatchCollection or PathCollection for speed.
This is what plt.scatter uses internally (it creates a PathCollection for all points).
"""

from matplotlib.collections import PatchCollection
from matplotlib.patches import Circle

plt.close()
fig, ax = plt.subplots()

# Create many small circle patches
circles = [Circle((i, i), 0.3*(i/5+1)) for i in range(5)]

# Add them as a collection (faster than adding individually)
collection = PatchCollection(circles, color="skyblue", alpha=0.6)
ax.add_collection(collection)

ax.set_xlim(-1, 6)
ax.set_ylim(-1, 6)
plt.show()
```

```python
#matplotlib.patches
"""
Defines individual geometric shapes (patches).
Examples: Circle, Rectangle, Polygon, Ellipse, Arc.
Each patch can be added to an Axes with ax.add_patch(...).
"""

from matplotlib.patches import Rectangle#, Circle -> already imported in previous cell

fig, ax = plt.subplots()

# Add shapes individually
ax.add_patch(Rectangle((1, 1), width=2, height=3, color="orange"))
ax.add_patch(Circle((4, 4), radius=1, color="blue", alpha=0.5))

ax.set_xlim(0, 6)
ax.set_ylim(0, 6)
plt.show()
```
